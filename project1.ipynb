```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Sales': [15000, 18000, 12000, 22000, 27000, 30000],
    'Profit': [4000, 5000, 3000, 7000, 9000, 10000]
}

df = pd.DataFrame(data)
```

```python
print("ðŸ“„ Sales Dataset:")
display(df)
```

```python
plt.figure(figsize=(8,5))
plt.plot(df['Month'], df['Sales'], color='royalblue', marker='o', label='Sales')
plt.plot(df['Month'], df['Profit'], color='seagreen', marker='s', label='Profit')
plt.title('Monthly Sales & Profit Trend')
plt.xlabel('Month')
plt.ylabel('Amount (â‚¹)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
best_month = df.loc[df['Sales'].idxmax(), 'Month']
print(f"ðŸ”¥ Highest Sales Month: {best_month}")
```

```python
average_profit = df['Profit'].mean()
print(f"ðŸ’° Average Monthly Profit: â‚¹{average_profit:.2f}")
```
